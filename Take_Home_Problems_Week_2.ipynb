1. Define two custom numpy arrays, say A and B. Generate two new numpy
arrays by stacking A and B vertically and horizontally.

```python
import numpy as np

```

```python
A = np.array(['a', 'b', 'c'])
B = np.array(['d', 'e', 'f'])

#stacking vertically
C = np.vstack((A,B))
print(C)

D = np.hstack((A,B))
print(D)

```

```python
A = np.array([['a'], ['b'], ['c']])
B = np.array([['d'], ['e'], ['f']])

#stacking vertically
C = np.vstack((A,B))
print(C)

D = np.hstack((A,B))
print(D)
```

4. Filter the rows of iris_2d that has petallength (3rd column) > 1.5
and sepallength (1st column) < 5.0


```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
# print(iris_2d)

indices = np.where((iris_2d[:,2]>1.5) & (iris_2d[:,0]<5.0))
print(iris_2d[indices])
```

6. Replace missing values in Min.Price and Max.Price columns with
their respective mean.

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

min_price_mean = df['Min.Price'].mean()
max_price_mean = df['Max.Price'].mean()
print(min_price_mean)
print(max_price_mean)

df['Min.Price'] = df['Min.Price'].fillna(min_price_mean)
df['Max.Price'] = df['Max.Price'].fillna(max_price_mean)

df.head()
```
